Importando a biblioteca pandas e o dados_hospedagem.json

```python
import pandas as pd
```

```python
dados = pd.read_json('/content/drive/MyDrive/Colab Notebooks/dados_hospedagem.json')
dados.head()
```

Os dados do json estavam aninhados. Usando o normalize para desestruturar esses dados.

```python
dados = pd.json_normalize(dados['info_moveis'])
dados
```

Visualizando as colunas como uma lista

```python
colunas = list(dados.columns)
colunas
```

```python
dados = dados.explode(colunas[3:])
# Explode os valores das colunas a partir do Ã­ndice 3 (ex: 'max_hospedes' em diante)
# Ou seja, se houver listas dentro dessas colunas, cada item da lista vira uma nova linha,
# mantendo o restante das informaÃ§Ãµes repetidas.
# obs: as listas precisam ter o mesmo tamanho, Ã© como se fosse um grande "for"
dados
```

```python
dados.reset_index(inplace=True, drop=True)
dados
```

Verificando o tipo de dados de dataframe

```python
dados.info()
```

Convertendo as colunas que possuem nÃºmeros inteiros e decimais para o tipo correto usando a biblioteca do numpy

```python
import numpy as np
```

```python
dados['max_hospedes'] = dados['max_hospedes'].astype(np.int64)
dados['max_hospedes']
```

```python
dados.info()
```

Ã‰ possÃ­vel converter mais de uma coluna ao mesmo tempo criando uma lista com o nome de colunas e a usando como o parÃ¢metro

```python
col_numericas = ['quantidade_banheiros', 'quantidade_quartos', 'quantidade_camas']
```

```python
dados[col_numericas] = dados[col_numericas].astype(np.int64)
```

```python
dados.info()
```

```python
dados['avaliacao_geral'] = dados['avaliacao_geral'].astype(np.float64)
```

```python
dados.info()
```

Tratando os valores monetÃ¡rios

```python
dados['preco']
```

Continuando o tratamento de valores monetÃ¡rios

```python
dados[['taxa_deposito', 'taxa_limpeza']]
```

Usado o applymap para conversÃ£o do dataframe

```python
dados[['taxa_deposito', 'taxa_limpeza']].applymap(lambda x : x.replace('$', '').replace(',', '').strip())
```

```python
dados[['taxa_deposito', 'taxa_limpeza']] = dados[['taxa_deposito', 'taxa_limpeza']].applymap(lambda x : x.replace('$', '').replace(',', '').strip())
```

```python
dados[['taxa_deposito', 'taxa_limpeza']] = dados[['taxa_deposito', 'taxa_limpeza']].astype(np.float64)
```
