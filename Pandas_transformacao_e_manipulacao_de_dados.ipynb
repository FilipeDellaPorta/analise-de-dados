Importando a biblioteca pandas e o dados_hospedagem.json

```python
import pandas as pd
```

```python
dados = pd.read_json('/content/drive/MyDrive/Colab Notebooks/dados_hospedagem.json')
dados.head()
```

Os dados do json estavam aninhados. Usando o normalize para desestruturar esses dados.

```python
dados = pd.json_normalize(dados['info_moveis'])
dados
```

Visualizando as colunas como uma lista

```python
colunas = list(dados.columns)
colunas
```

```python
dados = dados.explode(colunas[3:])
# Explode os valores das colunas a partir do índice 3 (ex: 'max_hospedes' em diante)
# Ou seja, se houver listas dentro dessas colunas, cada item da lista vira uma nova linha,
# mantendo o restante das informações repetidas.
# obs: as listas precisam ter o mesmo tamanho, é como se fosse um grande "for"
dados
```

```python
dados.reset_index(inplace=True, drop=True)
dados
```

Verificando o tipo de dados de dataframe

```python
dados.info()
```

Convertendo as colunas que possuem números inteiros e decimais para o tipo correto usando a biblioteca do numpy

```python
import numpy as np
```

```python
dados['max_hospedes'] = dados['max_hospedes'].astype(np.int64)
dados['max_hospedes']
```

```python
dados.info()
```

```python
dados[['taxa_deposito', 'taxa_limpeza']] = dados[['taxa_deposito', 'taxa_limpeza']].astype(np.float64)
```

Trabalhando com strings e tokenização

Transformando texto para minúsculo

```python
dados['descricao_local'] = dados['descricao_local'].str.lower()
dados['descricao_local']
```

```python
dados.head()
```

Removendo caracteres desnecessários com regex

```python
dados['descricao_local'][3169]
```

```python
dados['descricao_local'] = dados['descricao_local'].str.replace('[^a-zA-Z0-9\-\']', ' ', regex=True)
dados['descricao_local']
```

```python
dados['descricao_local'][3169]
```

```python
dados['descricao_local'] = dados['descricao_local'].str.replace('(?<!\w)-(?!\w)', ' ', regex=True) #removindo hífens que não componham palavras
```

```python
dados['descricao_local'] = dados['descricao_local'].str.split() #transformando em uma lista os elementos textuais separados por espaço
dados.head()
```

```python
dados.info()
```

```python
dados['comodidades'] = dados['comodidades'].str.replace('\{|}|\"', '', regex=True)
dados['comodidades']
```

```python
dados['comodidades'] = dados['comodidades'].str.split(',')
dados['comodidades']
```

```python
dados.info()
```

```python
dados.head()
```

```python
dados['descricao_vizinhanca']= dados['descricao_vizinhanca'].str.lower()
```

```python
dados['descricao_vizinhanca'][0]
```

```python
dados['descricao_vizinhanca'] = dados['descricao_vizinhanca'].str.replace('[^a-zA-Z0-9\-\']', ' ', regex=True).replace('(?<!\w)-(?!\w)', '', regex=True)
```
