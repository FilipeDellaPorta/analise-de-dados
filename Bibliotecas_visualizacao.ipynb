##Importando o pandas

```python
import pandas as pd
```

##Lendo e explorando os dados em csv

```python
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/imigrantes_canada.csv')
```

```python
df
```

```python
df.info()
```

##Trocando os índices para a coluna de 'País'

```python
df.set_index('País', inplace=True)
```

##Criando a variável anos para receber uma lista com números de 1980 a 2013 em formato de string

```python
anos = list(map(str, range(1980, 2014)))
```

```python
anos
```

##Filtrando o DataFrame para obter os dados do Brasil, retornando os anos e os respectivos números de imigrantes

```python
brasil = df.loc['Brasil', anos]
```

```python
brasil
```

##Foi criado um dicionário com as chaves 'ano' e 'imigrantes'. Os valores correspondem, respectivamente, às listas dos índices (anos de 1980 a 2013) e dos valores (quantidade de imigrantes). Em seguida, esse dicionário foi convertido em um DataFrame

```python
brasil_dict = {'ano': brasil.index.tolist(), 'imigrantes': brasil.values.tolist()}
dados_brasil = pd.DataFrame(brasil_dict)
```

```python
dados_brasil
```

##Importando pyplot e plotando gráfico

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8, 4)) #valores são em polegadas, primeiro largura e depois altura
plt.plot(dados_brasil['ano'], dados_brasil['imigrantes'])
plt.title('Imigração do Brasil para o Canadá')
plt.xlabel('Ano')
plt.ylabel('Nº de imigrantes')
plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010']) #melhorando a visualização dos labels do eixo x
plt.show() #removido as informações textuais da memória impressas #nenhuma linha depois é executada do plt.show
```

##Fazendo o mesmo com a Argentina (desafio) e comparando com o Brasil

```python
argentina = df.loc['Argentina', anos]
argentina_dict = {'ano': argentina.index.tolist(), 'imigrantes': argentina.values.tolist()}
dados_argentina = pd.DataFrame(argentina_dict)
plt.plot(dados_brasil['ano'], dados_brasil['imigrantes'], label='Brasil', color='green')
plt.plot(dados_argentina['ano'], dados_argentina['imigrantes'], label='Argentina', color='blue')
plt.legend()
plt.title('Imigração para o Canadá')
plt.xlabel('Ano')
plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010'])
plt.ylabel('Nº de imigrantes')
plt.grid(True)
plt.show()
```

##Criando uma figura

```python
fig, ax = plt.subplots(figsize=(8, 4))
ax.plot(dados_brasil['ano'], dados_brasil['imigrantes'])
ax.set_title('Imigração do Brasil para o Canadá\n1980 a 2013') #\n pula linha
ax.set_xlabel('Ano')
ax.set_ylabel('Nº de imigrantes')
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
plt.show()
```

##Plotando mais de um gráfico em uma mesma figura

```python
fig, axs = plt.subplots(1, 2, figsize=(15, 5)) #1 linha para dois gráficos um ao lado do outro #2 quantidade de gráficos
axs[0].plot(dados_brasil['ano'], dados_brasil['imigrantes'])
axs[0].set_title('Imigração do Brasil para o Canadá\n1980 a 2013')
axs[0].set_xlabel('Ano')
axs[0].set_ylabel('Nº de imigrantes')
axs[0].xaxis.set_major_locator(plt.MultipleLocator(5))
axs[0].grid()
axs[1].boxplot(dados_brasil['imigrantes'])
axs[1].set_title('Boxplot da imigração do Brasil para o Canadá\n1980 a 2013')
axs[1].set_xlabel('Brasil')
axs[1].set_ylabel('Nº de imigrantes')
axs[1].grid()
plt.show()
```

##Criando subplots em duas direções

```python
fig, axs = plt.subplots(2, 2, figsize=(10, 6))
fig.subplots_adjust(hspace=0.5, wspace=0.3)

axs[0, 0].plot(df.loc['Brasil', anos])
axs[0, 0].set_title('Brasil')

axs[0, 1].plot(df.loc['Colômbia', anos])
axs[0, 1].set_title('Colômbia')

axs[1, 0].plot(df.loc['Argentina', anos])
axs[1, 0].set_title('Argentina')

axs[1, 1].plot(df.loc['Peru', anos])
axs[1, 1].set_title('Peru')


for ax in axs.flat: #.flat transforma seu array 2D (ou de maior dimensão) em um iterador linear.
  ax.xaxis.set_major_locator(plt.MultipleLocator(5))
  ax.set_xlabel('Ano')
  ax.set_ylabel('Nº de imigrantes')

plt.show()
```

#Arrumando a distorção no eixo y

```python
fig, axs = plt.subplots(2, 2, figsize=(10, 6))
fig.subplots_adjust(hspace=0.5, wspace=0.3)
fig.suptitle('Imigração dos 4 maiores países da América do Sul para o Canadá\n1980 a 2013')

axs[0, 0].plot(df.loc['Brasil', anos])
axs[0, 0].set_title('Brasil')

axs[0, 1].plot(df.loc['Colômbia', anos])
axs[0, 1].set_title('Colômbia')

axs[1, 0].plot(df.loc['Argentina', anos])
axs[1, 0].set_title('Argentina')

axs[1, 1].plot(df.loc['Peru', anos])
axs[1, 1].set_title('Peru')


for ax in axs.flat:
  ax.xaxis.set_major_locator(plt.MultipleLocator(5))
  ax.set_xlabel('Ano')
  ax.set_ylabel('Nº de imigrantes')

ymin = 0
ymax = 7000

for ax in axs.ravel(): #axs.ravel() transforma a matriz de eixos em um array 1D, o que facilita fazer um for direto em todos os subplots — igual .flat, mas retornando um array em vez de um iterador.
  ax.set_ylim(ymin, ymax)

plt.show()
```
