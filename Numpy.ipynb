###Equação da reta:
y = ax + b

Entendi direitinho, Filipe. O que o instrutor está propondo é uma forma manual e intuitiva de ajustar uma reta aos dados, ou seja, tentar encontrar os coeficientes a (inclinação) e b (intercepto) da equação da reta y = ax + b por tentativa e erro, olhando o gráfico e testando valores até que a reta se aproxime visualmente dos pontos.

Mas sim, tem como descobrir os valores de forma matemática e automática: isso se chama regressão linear. A regressão linear encontra os melhores valores de a e b que minimizam os erros entre os pontos reais e a reta.

No Python (e no Colab), dá pra fazer isso rapidinho com bibliotecas como NumPy ou Scikit-Learn.

Quer que eu te mostre como fazer isso com NumPy? É bem direto.



```
import numpy as np
import matplotlib.pyplot as plt

# Suponha que esses são os dados (substitua pelos seus reais)
x = np.array([1, 2, 3, 4, 5])  # tempo (anos, meses, etc)
y = np.array([2.5, 3.0, 3.5, 4.0, 4.5])  # preços das maçãs

# Calculando os coeficientes da reta (a = inclinação, b = intercepto)
a, b = np.polyfit(x, y, 1)

print(f'Equação da reta: y = {a:.2f}x + {b:.2f}')

# Gerar os valores previstos com a reta
y_pred = a * x + b

# Plotando o gráfico
plt.scatter(x, y, color='red', label='Pontos reais')
plt.plot(x, y_pred, color='blue', label='Reta ajustada')
plt.xlabel('Tempo')
plt.ylabel('Preço da Maçã')
plt.legend()
plt.title('Ajuste de Reta com NumPy')
plt.grid(True)
plt.show()
```



````python
x = np.array(datas)
y = np.array(Moscow)
a, b = np.polyfit(x, y, 1) #equação da reta
print(f'Equação da reta: y = {a:.2f}x + {b:.2f}')
y_pred = a * x + b
plt.scatter(x, y, color='red', label='Pontos reais')
plt.plot(x, y_pred, color='blue', label='Reta ajustada')
plt.plot(datas, Moscow, label='Curva real')
plt.xlabel('Tempo')
plt.ylabel('Preço da Maçã')
plt.legend()
plt.title('Ajuste de Reta com NumPy')
plt.grid(True)
plt.show()
````

###Fazendo projeções (ex: data do 100º elemento):

````python
import numpy as np
import matplotlib.pyplot as plt

x = np.array(datas)
y = np.array(Moscow)

# Ajuste da reta
a, b = np.polyfit(x, y, 1)
print(f'Equação da reta: y = {a:.2f}x + {b:.2f}')

# Valores preditos com base na reta ajustada
y_pred = a * x + b

# Projeção para o ponto x = 100
x_proj = 100
y_proj = a * x_proj + b
print(f'Projeção para x = 100: y = {y_proj:.2f}')

# Gráfico
plt.scatter(x, y, color='red', label='Pontos reais')
plt.plot(x, y_pred, color='blue', label='Reta ajustada')
plt.plot(datas, Moscow, label='Curva real')

# Marcar a projeção
plt.scatter(x_proj, y_proj, color='green', s=100, label=f'Projeção (x=100)')
plt.axvline(x_proj, color='gray', linestyle='--', linewidth=1)

plt.xlabel('Tempo')
plt.ylabel('Preço da Maçã')
plt.legend()
plt.title('Ajuste de Reta com Projeção')
plt.grid(True)
plt.show()
````
